```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
from statsmodels.tsa.arima_process import arma_generate_sample
```

## 1. AR Process

```python
np.random.seed(12345)
```

```python
ar1 = arma_generate_sample(
    ar=np.array([1, 0.5]),
    ma=np.array([1]),
    scale=1,
    nsample=100,
    burnin=10
)
```

```python
mu_ar1 = np.mean(ar1)
```

```python
fig_ar1 = plt.figure(figsize=(12, 6))
ax_ar1 = fig_ar1.add_subplot(111)
ax_ar1.plot(range(100), ar1, label="DGP")
ax_ar1.grid(True)
ax_ar1.axhline(mu_ar1, color="r", linestyle="dashed", label="full-sample mean")
ax_ar1.legend()
ax_ar1.set_title("AR(1) process sample")
```
